# Algoritmo K-Means

# ¿Que es?

El algoritmo k-means es un método de agrupamiento que divide un conjunto de datos en k grupos o clusters. Los datos se agrupan de tal manera que los puntos en el mismo clúster sean más similares entre sí que los puntos en otros clusters.

# ¿Como Funciona?


1.   Especificar el número de clústers deseados (k)
2.   Seleccionar k puntos al azar del conjunto de datos como los centroides iniciales de cada clúster
3.   Asignar cada punto del conjunto de datos al cluster cuyo centroide esté más cerca
4.   Recalcular los centroides de cada cluster como la media de todos los puntos del cluster
5.   Repetir los pasos 3 y 4

````python

listaX=[]; #almacena los valores de las coordenadas X
listaY=[]; #almacena los valores de las coordenadas Y
listCentroidesX=[] #almacena los valores de las coordenas X del centroide
listCentroidesY=[] #almacena los valores de las coordenas Y del centroide
nuevoCentroideX=[] #almacena los valores de las coordenas X del nuevo centroide
nuevoCentroideY=[] #almacena los valores de las coordenas Y del nuevo centroide
bandera=False #bandera que determina si el los valores del nuevo centroide coinciden con el anterior
````

````python
import math

listaX=[]; #almacena los valores de las coordenadas X
listaY=[]; #almacena los valores de las coordenadas Y
listCentroidesX=[] #almacena los valores de las coordenas X del centroide
listCentroidesY=[] #almacena los valores de las coordenas Y del centroide
nuevoCentroideX=[] #almacena los valores de las coordenas X del nuevo centroide
nuevoCentroideY=[] #almacena los valores de las coordenas Y del nuevo centroide
bandera=False #bandera que determina si el los valores del nuevo centroide coinciden con el anterior

# Buble FOR que llenan las coordenas X e Y
for i in range(1,7):
    listaX.insert(i,input("Introduce la coordenada"+str(i)+" (X,"))
    listaY.insert(i,input("Introduce la coordenada"+str(i)+" ,Y)"))

# Bucle For que ordena las coordenadas X en orden Ascendente
for i in range(len(listaX)-1):
        for j in range(len(listaX)-1-i):
            if listaX[j] > listaX[j+1]:
                listaX[j], listaX[j+1] = listaX[j+1], listaX[j]
                listaY[j], listaY[j+1] = listaY[j+1], listaY[j]

totCentroides=int(input("\n Define el número de centroides: "))

# Bucle FOR que obtiene las coordenadas X e Y segun la cantidad de centroides
for i in range(totCentroides):
    listCentroidesX.insert(i,listaX[i])
    listCentroidesY.insert(i,listaY[i])
    nuevoCentroideX.insert(i,0)
    nuevoCentroideY.insert(i,0)


# imprime los valores de los puntos X e Y así como las coordenadas del centroide
print("*********************************")
print("Puntos con coordenadas X,Y")
print(listaX)
print(listaY)
print("*********************************")
print("Centroides base con cordenadas X,Y")
print(listCentroidesX)
print(listCentroidesY)

# Ciclo WHILE que se repite N veces hasta que los nuevos centroides coincidan con los anteriores.
while bandera!=True:
    puntos="\n las coordenadas y valores de los centroides son: \n"
    resCoordenadas=0 #almacena el valor de cada punto tras aplicar la fórmula

    matrizNumCentroide=[] # Matriz que lleva el registro de cada centroide con su respectiva coordenada
    matrizValPuntos=[]    # Matriz que almacena el valor de cada punto, servirá para comparar a que centroide pertenece
    matrizCoordenadaX=[]  # Matriz que almacena la coordenada X del punto
    matrizCoordenadaY=[]  # Matriz que almacena la coordenada Y del punto

    #Ciclo FOR que genera el valor de cada punto para cada centroide con sus respectivas coordenadas
    for i in range(len(listCentroidesX)):
        listaValCentroide=[]                # Lista que almacena el valor de cada punto de cada centroide
        listaNumCentroide=[]                # Lista que almacena el número del centroide
        listaNum_coordenada_CentroideX=[]   # Lista que almacena la coordenada X del centroide
        listaNum_coordenada_CentroideY=[]   # Lista que almacena la coordenada Y del centroide

        # Por cada vuelta se vacía la lista para guardar los valores nuevos
        listaValCentroide.clear()
        listaNumCentroide.clear()
        listaNum_coordenada_CentroideX.clear()
        listaNum_coordenada_CentroideY.clear()

        # Ciclo FOR interno que recorre cada punto para cada centroide
        for j in range(len(listaX)): # obtiene el total de coordenadas ingresadas
            # Aplica la formula para obtener el valor de los puntos
            resCoordenadas=round(math.sqrt((math.pow((float(listaX[j])-float(listCentroidesX[i])),2))+(math.pow((float(listaY[j])-float(listCentroidesY[i])),2))),2)
            listaValCentroide.insert(j,resCoordenadas)       # En cada posición de la lista guarda el valor de cada punto
            listaNumCentroide.append(i+1)                    # Guarda el número de centroide que le corresponde
            listaNum_coordenada_CentroideX.append(listaX[j]) # Guarda la coordenada X del centroide
            listaNum_coordenada_CentroideY.append(listaY[j]) # Guarda la coordenada Y del centroide

            #Concatena los valores anteriores para visualizar de forma mas real las coordenadas con el valor de los puntos
            puntos+="("+str(listCentroidesX[i])+","+str(listCentroidesY[i])+")("+str(listaX[j])+","+str(listaY[j])+")="+str(resCoordenadas)+"\n"
        matrizValPuntos.append(listaValCentroide)                # Por cada vuelta interna se genera una lista de valores de puntos y se guarda en la matriz
        matrizNumCentroide.append(listaNumCentroide)             # Por cada vuelta interna se genera una lista de Número de centroide y se guarda en la matriz
        matrizCoordenadaX.append(listaNum_coordenada_CentroideX) # Por cada vuelta interna se guarda en la lista la coordena X del centroide y se guarda en la matriz
        matrizCoordenadaY.append(listaNum_coordenada_CentroideY) # Por cada vuelta interna se guarda en la lista la coordena Y del centroide y se guarda en la matriz

    # Se imprime la cadena concatenada
    print("*********************************")
    print(puntos)

    resMatrizNCentroide=[] # Matriz que almacena el número de cada centroide tras comparar que centroide se queda con el punto
    resMatrizCentroideX=[] # Matriz que almacena la coordenada X del punto tras comparar el valor
    resMatrizCentroideY=[] # Matriz que almacena la coordenada Y del punto tras comparar el valor

    #coordenada=0
    esMenorAux=0     # Variable que guarda el valor de la primera posición de la matriz para ir comparando con los demás puntos
    numCentroide=0   # Variable que guarda el número del centroide tras comparar que valor es menor
    coorCentroideX=0 # Variable que guarda el valor de la coordenada X del centroide tras comparar que valor es menor
    coorCentroideY=0 # Variable que guarda el valor de la coordenada Y del centroide tras comparar que valor es menor
    #concatNumCentroide="centroide"

    # Bucle FOR que sirve para comparar el valor menor de cada punto y asignar al centroide correspondiente
    for i in range(len(listaX)): #obtiene tamaño del total de las coordenas introducidas al incio y recorre cada columna
        esMenorAux=matrizValPuntos[0][i] # obtiene el primer valor a comparar
        numCentroide=matrizNumCentroide[0][i]   # Obtiene el primer valor del número de centroide
        coorCentroideX=matrizCoordenadaX[0][i]  # Obtiene el primer valor de la coordenada X del centroide
        coorCentroideY=matrizCoordenadaY[0][i]  # Obtiene el primer valor de la coordenada Y del centroide
        for j in range(len(matrizValPuntos)-1): # recorre cada fila de la matriz que tiene los
            if(esMenorAux>matrizValPuntos[j+1][i]):      # Realiza la comparación para determinar que centroide se queda con la coordenada
                esMenorAux=matrizValPuntos[j+1][i]       # Si el siguiente punto es menor entonces se le asigna a la variable
                numCentroide=matrizNumCentroide[j+1][i]  # Se guarda el número del centroide que se queda con la coordenada
                coorCentroideX=matrizCoordenadaX[j+1][i] # Se guarda la coordenada X del centroide ganador
                coorCentroideY=matrizCoordenadaY[j+1][i] # Se guarda la coordenada Y del centroide ganador
        resMatrizNCentroide.append(numCentroide)         # de cada lista generada se almacena el número del centroides ganadores
        resMatrizCentroideX.append(coorCentroideX)       # de cada lista generada se almacena en la coordenada X del centroide ganador
        resMatrizCentroideY.append(coorCentroideY)       # de cada lista generada se almacena en la coordenada Y del centroide ganador

    # Una vez obtenido las coordenadas de cada centroide ganador,
    # se ordenan por número de centroide para poder sumar cada coordenada
    for i in range(len(resMatrizNCentroide)-1):
        for j in range(len(resMatrizNCentroide)-1-i):
            if resMatrizNCentroide[j] > resMatrizNCentroide[j+1]:
                resMatrizNCentroide[j], resMatrizNCentroide[j+1] = resMatrizNCentroide[j+1], resMatrizNCentroide[j]
                resMatrizCentroideX[j], resMatrizCentroideX[j+1] = resMatrizCentroideX[j+1], resMatrizCentroideX[j]
                resMatrizCentroideY[j], resMatrizCentroideY[j+1] = resMatrizCentroideY[j+1], resMatrizCentroideY[j]

    # Muestra cada centroide con sus respectivas coordenadas
    print("*********************************")
    print("Centroides con sus respectivas coordenadas:")
    print(resMatrizNCentroide)
    print(resMatrizCentroideX)
    print(resMatrizCentroideY)

    matrizCoordenadaCentroideX=[] # Matriz que sirve para agrupar las coordenas X de cada centroide
    matrizCoordenadaCentroideY=[] # Matriz que sirve para agrupar las coordenas Y de cada centroide

    # Bucle FOR que agrupa las coordenas x e Y de cada centroide para generar las nuevas coordenadas de los centroides
    for i in range(totCentroides):
        arrayCentroidesX=[]
        arrayCentroidesX.clear()
        arrayCentroidesY=[]
        arrayCentroidesY.clear()
        for j in range(len(resMatrizNCentroide)):
            if((i+1)==resMatrizNCentroide[j]):
                arrayCentroidesX.insert(j,resMatrizCentroideX[j])
                arrayCentroidesY.insert(j,resMatrizCentroideY[j])
        matrizCoordenadaCentroideX.append(arrayCentroidesX)
        matrizCoordenadaCentroideY.append(arrayCentroidesY)

    # Imprime las listas agrupadas con sus respectivas corrdenadas X e Y de cada centroide
    print("\n*********************************")
    print("Matriz de coordenadas X, Y por de cada Centroide")
    print(matrizCoordenadaCentroideX)
    print(matrizCoordenadaCentroideY)

    # Como cada centroide tiene diferente número de coordenadas, se obtiene el tamaño de cada lista
    nuevoCentroideX.clear() # Por cada vuelta del ciclo WHILE se limpian las coordenadas X del centroide
    nuevoCentroideY.clear() # Por cada vuelta del ciclo WHILE se limpian las coordenadas Y del centroide

    # Bucle for que realiza la suma de las coordenadas X e Y de cada centroide y los divide entre el total de coordenadas ganadas
    for i in range(len(matrizCoordenadaCentroideX)): #toma el tamaño de la matriz, es decir el total de filas
        sumaCoordenadaX=float(matrizCoordenadaCentroideX[i][0]);
        sumaCoordenadaY=float(matrizCoordenadaCentroideY[i][0]);
        contDiv=int(1);
        for j in range(len(matrizCoordenadaCentroideX[i])-1): #toma el tamaño de columnas de cada fila de la matriz
            contDiv+=1;
            sumaCoordenadaX+=float(matrizCoordenadaCentroideX[i][j+1])
            sumaCoordenadaY+=float(matrizCoordenadaCentroideY[i][j+1])
        nuevoCentroideX.insert(i,round((sumaCoordenadaX/contDiv),1)) # Asigna la nueva coordenada X del nuevo centroide
        nuevoCentroideY.insert(i,round((sumaCoordenadaY/contDiv),1)) # Asigna la nueva coordenada Y del nuevo centroide

    # Imprime los nuevos centroides
    print("\n*********************************")
    print("Nuevos centroides:")
    print(nuevoCentroideX)
    print(nuevoCentroideY)


    arrayBandera=[] # Almacena un valor boleano resultante de la comparación

    # Bucle FOR que comparar el valor de cada coordenada X e Y de los centroides para determinar si son iguales
    for i in range(len(nuevoCentroideX)):
        if(listCentroidesX[i]==nuevoCentroideX[i] and listCentroidesY[i]==nuevoCentroideY[i]):
            arrayBandera.insert(i,True)
        else:
            listCentroidesX[i]=nuevoCentroideX[i]
            listCentroidesY[i]=nuevoCentroideY[i]
            arrayBandera.insert(i,False)

    # Una vez llenado la lista de banderas, se recorre para verificar que todos sean verdaderos
    # En caso de que una de las comparaciones de False, significa que los centroides no son iguales
    # Y se debe repetir el ciclo

    for i in range(len(arrayBandera)-1):
        bandera=arrayBandera[i] and arrayBandera[i+1]; # Compara el valor de cada celda con la siguiente
    if(bandera==True):
        print("\n\n********Felicidades haz llegado al final del algoritmo********!!")
    else:
        print("\n**Sigue calculando!!")      }
````
